```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la función
def f(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

# Rango de valores para x'
x_prime = np.linspace(0, 1, 1000)

# Calculamos los valores de la función
y = f(x_prime)

# Configuración del gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_prime, y, 'b-', linewidth=2)
plt.title('Perturbación inicial en la cuerda\n$f(x\') = \exp\\left(-\\frac{(x\' - 0.5)^2}{0.01}\\right)$')
plt.xlabel('Posición (x\')')
plt.ylabel('Amplitud de la perturbación')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 1)
plt.ylim(0, 1.1)

# Destacamos el centro de la gaussiana
plt.axvline(x=0.5, color='r', linestyle='--', alpha=0.5)
plt.text(0.5, -0.1, 'x\' = 0.5', ha='center', va='top', color='r')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Reutilizamos la función original
def f(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

# Configuración de parámetros
x = np.linspace(0, 2, 1000)  # Ampliamos el rango para ver el movimiento
c = 0.5  # Aumentamos la velocidad para mejor visualización
t_values = np.linspace(0, 2, 50)  # Tiempo de 0 a 2 segundos

# Configuración del gráfico
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, 2)
ax.set_ylim(0, 1.1)
ax.set_xlabel('Posición (x)')
ax.set_ylabel('Amplitud')
ax.set_title('Inciso (a): Perturbación moviéndose a la derecha\n$x\' = x - ct$')
ax.grid(True, linestyle='--', alpha=0.5)

# Elementos del gráfico
line, = ax.plot([], [], 'b-', lw=2)
time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)
center_text = ax.text(0.02, 0.85, '', transform=ax.transAxes, color='red')

# Función de inicialización
def init():
    line.set_data([], [])
    time_text.set_text('')
    center_text.set_text('')
    return line, time_text, center_text

# Función de animación
def animate(t):
    x_prime = x - c*t  # Aplicamos el cambio de variable
    y = f(x_prime)
    line.set_data(x, y)

    # Actualizamos textos
    time_text.set_text(f'Tiempo: {t:.2f} s')
    center_pos = 0.5 + c*t  # Calculamos la nueva posición del centro
    center_text.set_text(f'Centro en x = {center_pos:.2f}')

    return line, time_text, center_text

# Creamos la animación
anim = FuncAnimation(fig, animate, frames=t_values,
                     init_func=init, interval=100, blit=True)

plt.close()
HTML(anim.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Definición de la función de perturbación
def f(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

# Configuración del espacio y tiempo
x = np.linspace(0, 2, 1000)  # Rango espacial extendido
c = 0.1  # Velocidad de propagación
t_values = np.linspace(0, 5, 100)  # Valores de tiempo

# Configuración de la figura
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xlim(0, 2)
ax.set_ylim(-1.2, 1.2)
ax.set_xlabel('Posición (x)')
ax.set_ylabel('Amplitud')
ax.set_title('Superposición de Ondas Viajeras (Inciso c)')
ax.grid(True, linestyle='--', alpha=0.5)

# Líneas para las ondas individuales y la superposición
line_right, = ax.plot([], [], 'b-', lw=2, alpha=0.5, label='Onda derecha (a)')
line_left, = ax.plot([], [], 'r-', lw=2, alpha=0.5, label='Onda izquierda (b)')
line_super, = ax.plot([], [], 'k-', lw=3, label='Superposición')
time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)
ax.legend(loc='upper right')

# Función de inicialización
def init():
    line_right.set_data([], [])
    line_left.set_data([], [])
    line_super.set_data([], [])
    time_text.set_text('')
    return line_right, line_left, line_super, time_text

# Función de animación
def animate(t):
    # Onda moviéndose a la derecha (inciso a)
    x_prime_right = x - c*t
    y_right = f(x_prime_right)
    line_right.set_data(x, y_right)

    # Onda moviéndose a la izquierda (inciso b)
    x_prime_left = x + c*t
    y_left = f(x_prime_left)
    line_left.set_data(x, y_left)

    # Superposición (inciso c)
    y_super = y_right + y_left
    line_super.set_data(x, y_super)

    time_text.set_text(f'Tiempo: {t:.2f} s')
    return line_right, line_left, line_super, time_text

# Crear la animación
anim = FuncAnimation(fig, animate, frames=t_values,
                    init_func=init, interval=100, blit=True)

plt.close()
HTML(anim.to_jshtml())
```
