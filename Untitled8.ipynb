```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Pulso unitario exacto
def f_pulso(x):
    return np.where(np.abs(x) <= 1, 1, 0)

# Aproximación g(x) con integración numérica hasta cierto límite
def g_approx(x, upper_limit):
    result, _ = quad(lambda w: (np.sin(w) / w) * np.cos(w * x), 0, upper_limit, limit=200)
    return (2 / np.pi) * result

# Aproximación exacta g(x) para integral hasta ∞
def g_exact(x):
    return np.where(np.abs(x) < 1, 1,
           np.where(np.abs(x) == 1, 0.5, 0))

# Vectorizadores
def g_vectorized(upper_limit):
    return np.vectorize(lambda x: g_approx(x, upper_limit))

# Rango de x
x_vals = np.linspace(-5, 5, 500)
f_vals = f_pulso(x_vals)

# Aproximaciones numéricas
limits = [8, 16, 32]
colors = ['blue', 'green', 'red']
g_vals_list = [g_vectorized(limit)(x_vals) for limit in limits]

# Aproximación exacta (∞)
g_inf_vals = g_exact(x_vals)

# --- Graficar ---
plt.figure(figsize=(12, 6))
plt.plot(x_vals, f_vals, '--', label='Pulso unitario $f(x)$', linewidth=2, color='black')

# Aproximaciones numéricas
for i, limit in enumerate(limits):
    plt.plot(x_vals, g_vals_list[i], label=fr'Aprox $g(x)$ con $\int_0^{{{limit}}}$', linewidth=2, color=colors[i])

# Curva exacta (integral hasta ∞)
plt.plot(x_vals, g_inf_vals, label=r'$g(x)$ exacta $\int_0^\infty$', linewidth=2, color='orange', linestyle=':')

plt.title('Reconstrucción del pulso unitario con integrales truncadas y exacta')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
