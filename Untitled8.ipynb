Solucin.\
Resolveremos el ejercicio con el mtodo de separacin de variables.\
Sea $U(x,y,t)=F(x,y)G(t)$, usando la ecuacin de onda, tenemos:\
$$FG''=c^2(F_{xx}G+F_{yy}G)$$
$$\frac{G''}{c^2G}=\frac{1}{F}(F_{xx}+F_{yy})$$
Tenemos que ambos miembros deben ser iguales a una constante(Problema de Sturm-Lioville):\
Asi, sabemos que tienen solucin con $-\omega^2$:\
$$\frac{G''}{c^2G}=\frac{1}{F}(F_{xx}+F_{yy})=-\omega^2$$
Para $G(t):$
$$G''+\lambda G=0, \quad con \lambda=c\omega$$
Y para la "funcin de amplitud" $F(x,y)$ se tiene la E.D.P.
$$F_{xx}+F_{yy}+\omega^2F=0$$
Usando la separacin de la ecuacin de Helmholtz:
$$F(x,y)=H(x)Q(y)$$
Derivando para sustituir en la funcin de amplitud nos queda:
$$\frac{d^2H}{dx^2}Q=-\left(H\frac{d^2Q}{dy^2}+\omega^2 HQ\right)$$
Separando las variables:
$$\frac{1}{H} \frac{d^2H}{dx^2}=\frac{1}{Q}$$

$$\frac{d^2H}{dx^2}Q=-\left(H\frac{d^2Q}{dy^2}+\omega^2HQ \right)$$\
$$\frac{1}{H}\frac{d^2H}{dx^2}=-\frac{1}{Q}\left(H\frac{d^2Q}{dy^2}+\omega^2Q \right)= -k^2$$ como es un problema de sturm tiene que estar igualado a una constante k\
Por lo tanto para la funcin H tenemos que $$\frac{d^2H}{dx^2}+k^2H=0$$
y para Q
$$\frac{d^2Q}{dy^2}+p^2Q=0$$
donde $p^2=\omega^2-k^2$



Aplicando las condiciones de frontera:
$H(x)=Acos(kx)+Bsen(kx)$ y $Q(y)=C cos(py)+Dsen(py)$, con A,B,C,D constantes.
Ahora, Con la condicin de frontera $U=0 \quad\forall t \geq 0 $ entonces:
$$F=HQ=0 \implies x=0,x=a,y=0,y=b$$
asi:
$$H(0)=0,\quad H(a)=0, \quad Q(0)=0, \quad Q(b)=0$$
Usando la tabla dada tenemos:
$$\lambda=\lambda_{mn}=c\pi \sqrt{\frac{m^2}{a^2}+\frac{n^2}{b^2}}, \quad m,n=1,2,3,...$$
Usando $G_{mn}=B_{mn}cos(\lambda_{mn}t)+B^* sen(\lambda_{mn}t)$\
Asi:
$$U_{mn}(x,y,t)=(B_{mn}cos(\lambda_{mn}t)+B^* sen(\lambda_{mn}t))sen(\frac{m\pi x}{a})sen(\frac{m\pi y}{b})$$

Para determinar los valores de $B_{mn}$ y $B^*_{mn}$:


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

# Parmetros del problema
a = 4  # longitud en x
b = 2  # longitud en y
c = 1  # velocidad de propagacin

# Nmero de trminos en la serie doble de Fourier
N = 10
M = 10

# Funcin inicial f(x,y)
def f(x, y):
    return 0.1 * (4*x - x**2)*(2*y - y**2)

# Dominio
Lx = a
Ly = b

# Resolucin espacial y temporal
nx = 50
ny = 50
x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)
X, Y = np.meshgrid(x, y)

# Coeficientes de Fourier doble
def B_mn(m, n):
    integrand = f(X, Y) * np.sin(m * np.pi * X / Lx) * np.sin(n * np.pi * Y / Ly)
    dx = Lx / (nx - 1)
    dy = Ly / (ny - 1)
    integral = np.sum(integrand) * dx * dy
    return (4 / (Lx * Ly)) * integral

# Solucin u(x, y, t)
def u(x, y, t):
    result = np.zeros_like(x)
    for m in range(1, N+1):
        for n in range(1, M+1):
            B = B_mn(m, n)
            omega_mn = c * np.pi * np.sqrt((m/Lx)**2 + (n/Ly)**2)
            term = (B * np.cos(omega_mn * t) *
                    np.sin(m * np.pi * x / Lx) *
                    np.sin(n * np.pi * y / Ly))
            result += term
    return result

# Visualizacin en t = 0, t = 0.5, t = 1.0
times = [0, 0.5, 1.0]
fig = plt.figure(figsize=(18, 5))

for i, t in enumerate(times):
    ax = fig.add_subplot(1, 3, i+1, projection='3d')
    U = u(X, Y, t)
    surf = ax.plot_surface(X, Y, U, cmap=cm.viridis)
    ax.set_title(f'u(x, y, t={t})')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('u')
    ax.set_zlim(-0.5, 0.5)

plt.tight_layout()
plt.show()
```
