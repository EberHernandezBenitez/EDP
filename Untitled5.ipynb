Calculando la serie de Fourier generalizada usando lo antes calculado para $U(x)=e^{-x}$, 0<=x<=1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from math import pi, exp
from pylab import *
def exponential(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = np.exp(-x[i])
    return y
```

```python
x = linspace(0,1,10)
f = exponential(x)
plot(x,f,label='$u(x,0)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-0.01,1.01]),ylim([-0.01,1.01]),grid(),legend()
show()
```

```python
def u_0(x,N=200): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += 2*(1-(-1)**n)*(np.sin(n*pi*x))/n
    return (1/pi)*f
```

```python
f_n = u_0(x)
plot(x,f_n*f,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-0.01,1.1]),ylim([-0.01,1.2]),grid(),legend()
show()
```
