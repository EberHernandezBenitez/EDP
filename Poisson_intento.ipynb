# Generación de una variable aleatoria Poisson.
Una variable $X$ de Poisson con media $\lambda$ si:
$$p_i=P(X=i)=\frac{e^{-\lambda}\lambda^i}{i!}, \quad i=0,1,2,....$$
Con ello, aplicaremos el seudocódigo dado en clase pero necesitamos entender lo que representa esta variable aleatoria.

La distribución de Poisson es un modelo matemático que calcula la probabilidad de que ocurra un número determinado de eventos independientes en un intervalo fijo de tiempo o espacio, conociendo de antemano su tasa promedio de ocurrencia ($\lambda$). \
Lambda nos indica el número promedio historico de eventos y la fórmula nos da la probabilidad de que ocurra $i$ eventos en el intervalo de tiempo dado.\
Ahora, el seudocódigo no calcula una probabilidad, sino que simula un escenario real, genera un número entero aleatorio ($X \ge 0$) que se comporta exactamente como si estuviera ocurriendo en la vida real bajo una distribución de Poisson. \
Y una manera de verificar que el número generado respete la probabilidad de Poisson usamos el método de la transformadad inversa.

Para el seudocódigo, usamos lo visto en clase:

```python
import math
import random
def f(lam):
    U = random.random()
    i = 0
    p = math.exp(-lam)  # p_0 = e^(-lambda)
    F = p  # F(0) = p_0
    while True:
        if U < F:
            return i  # Devuelve X = i y detiene la función

        # Actualización usando la relación de recurrencia p_i+1
        p = (lam * p) / (i + 1)
        F = F + p
        i = i + 1
```
