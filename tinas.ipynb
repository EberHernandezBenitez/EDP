La empresa TIBASA (Fabricante de tinas de baño) tiene asignado un camión especial para el transporte de tinas terminadas. Dicho camión transporta diariamente 5 tinas. El peso de cada tina sigue la distribución triangular de la imagen anexa.

Si la capacidad del camión es de 1 tonelada, ¿cuál es la probabilidad de que el peso de las tinas exceda la capacidad del camión?

**PASO 1**. Defina las variables que intervienen en el modelo.
En base a la grafica dada podemos observar que sigue una distribución triangular.\
Peso minimo de cada tina **a** = 190 kg\
Peso máximo de cada tina **c** = 230 kg\
El peso "moda" **b** = 210 kg\
Capacidad del camión: 1000 kg\
Número de tinas por viaje: 5 (n=5)\
Como variable $x_i=$ peso de cada tina. \
$x$ SIGUE UNA DISTRIBUCIÓN TRIANGULAR.


**Paso 2.** Sabemos que la función sigue una distribución triangular, asi:\
$$f(x) =
\begin{cases}
0 & x < 190 \\
\frac{x - 190}{200} & 190 \le x \le 210 \\
\frac{230 - x}{200} & 210 < x \le 230 \\
0 & x > 230
\end{cases}$$

**Paso 3**: La distribución acumulada de probabilidad de cada variable.\
$$F(x) = \begin{cases}
0 & x < 190 \\
\frac{(x-190)^2}{(230-190)(210-190)} & 190 \le x \le 210 \\
1 - \frac{(230-x)^2}{(230-190)(230-210)} & 210 < x \le 230 \\
1 & x > 230
\end{cases}$$

**Paso 4**:Buscamos la función inversa con $F(x)=R$ y el punto de inflexión:
En $F(x=210)$ $R=0.5$ asi:
$$x = \begin{cases}
\sqrt{800R}+190 & 0 \le R \le 0.5 \\
230 - \sqrt{800R} & 0.5 < R \le 1
\end{cases}$$

**Paso 5:** Crear el seudocódigo

```python
import random as r
import numpy as np
import matplotlib.pyplot as plt

def bañera(n):
  l=[]
  exceso=0
  for i in range(n):
    u = r.random()
    if 0 <= u < 0.5:
      x= np.sqrt(800*u)+190
      l.append(x)
    else:
      x = 230- np.sqrt(800*u)
      l.append(x)
  return l
  if l <= 1000:
    exceso += 0
  else:
    exceso +=1
```

```python
bañera(5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
a = 190  # Mínimo
c = 210  # Moda
b = 230  # Máximo

# Parámetros
tinas = 5
capacidad_camion = 1000
simulaciones = 100000

def simtinas(n_sim, tinas, a, c, b, capacidad):
    pesos = np.random.triangular(a, c, b, size=(n_sim,tinas))

    # Sumamos los pesos de las 5 tinas
    pesosviaje = np.sum(pesos, axis=1)

    # calcular los excesos
    excesos = np.sum(pesosviaje > capacidad)

    # Calculamos la probabilidad
    probabilidad = excesos / n_sim

    return probabilidad, pesosviaje

# Ejecutar simulación
prob, resultados = simtinas(simulaciones, tinas, a, c, b, capacidad_camion)

print(f"--- Resultados de la Simulación ---")
print(f"Probabilidad de exceder 1000kg: {prob:.4f} ({prob*100:.2f}%)")
print(f"Peso promedio simulado: {np.mean(resultados):.2f} kg")
```
