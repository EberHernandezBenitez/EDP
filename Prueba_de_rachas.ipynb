```python
def racha(D):
  signos=[]

  conteos =1
  for i in range(len(D)-1):
    if D[i+1] < D[i]:
      signos.append(0)
    else:
      signos.append(1)
  for i in range(len(signos)-1):
    if signos[i] != signos[i+1]:
      conteos +=1
    else:
      conteos +=0
  #calculando ahora mu, sigma y el estadistico de prueba
  alfa=0.05
  z_n = stats.norm.ppf(1-alfa/2)
  n=len(datos)
  media= (2*n-1)/3
  sigma= (16*n-29)/90
  estadistico= (conteos-media)/sigma
  print (f"el nùmero de conteos es {conteos}")
  print(f"el estadistico de prueba es {estadistico}")
  print(f"la media es {media}")
  print(f"el valor critico es {z_n}")
  print(f"sigma es {sigma}")
  if estadistico < abs(z_n) :
    print("no se rechaza la hipotesis nula")
  else:
    print("se rechaza la hipotesis nula")
  return
```
