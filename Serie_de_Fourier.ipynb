```python
from pylab import *
```

Importamos de pylab las librerias para graficar


```python
def rect(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = 2 - x[i]
    return y
def T(a):
    y = zeros(len(a))
    for i in range(len(a)):
        y[i] = 0
    return y
```

Definimos la funcion a trazos la funcion rec(x) es para crear la parte de recta de 2-x, y la grafica g(a)=0 cuando 0


















































































































































































































































































































































```python
def fourier(x1,N=10):
    J = zeros(len(x1))
    for n in range(1,N):
        J += ((((cos(n*pi*x1/2))*(1-((-1)**n))/((n**2)*(pi))))+((sin(n*pi*x1/2))/n))
    return 1/2 + (2.0/pi)*J
```

Creamos la funcion para la serie de fourier

```python
x= linspace(0,2,1000)
f = rect(x)
a = linspace(-2,0,1000)
g = T(a)
x1= linspace(-2,2,1000)
E= fourier(x1)
plot(x,f,label='$f(x)= 0$')
plot(a,g,label='$f(x)= 2-x $')
plot(x1,E,label='Serie de Fourier')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Fúnción a trozos")
xlim([-2.00,2.00]),ylim([-2.0,2.00]),grid(),legend()

show()
```
