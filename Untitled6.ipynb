Para graficar los polinomios de Legrende(inciso a).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

# Dominio para graficar
x_vals = np.linspace(-1, 1, 400)
plt.figure(figsize=(10, 6))

# Graficando P_{n}:
for n in range(6):
    Pn = legendre(n)
    y_vals = Pn(x_vals)  # Evalua en x
    plt.plot(x_vals, y_vals, label=f'$P_{n}(x)$')

#Para la grafica:
plt.title('Polinomios de Legendre $P_n(x)$ para $n=0,..,5$')
plt.xlabel('$x$')
plt.ylabel('$P_n(x)$')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

Para el inciso b necesitamos ver que se tienen las mismas graficas de ambos lados de la igualdad.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

# Parmetro
n = 3
x_vals = np.linspace(-1, 1, 400)

# Polinomios P_n(x) y P_{n+1}(x)
Pn = legendre(n)
Pn1 = legendre(n + 1)

# Para la derivada:
Pn_deriv = Pn.deriv()

# Lado izquierdo: (1 - x^2) * Pn'(x)
lhs_vals = (1 - x_vals**2) * Pn_deriv(x_vals)

# Lado derecho: (n+1)x*Pn(x) - (n+1)*Pn1(x)
rhs_vals = (n + 1) * x_vals * Pn(x_vals) - (n + 1) * Pn1(x_vals)

# Para la grafica de ambos lados:
plt.figure(figsize=(10, 6))
plt.plot(x_vals, lhs_vals, label='Lado Izquierdo: $(1 - x^2) P_3\'(x)$', linestyle='--',color='black')
plt.plot(x_vals, rhs_vals, label='Lado Derecho: $4xP_3(x) - 4P_4(x)$', linestyle=':',color='pink')

plt.title('Verificacin grfica para $n = 3$')
plt.xlabel('$x$')
plt.legend()
plt.grid(True)
plt.axhline(0, color='green', lw=0.5)
plt.axvline(0, color='green', lw=0.5)
plt.show()
```

Inciso c, necesitamos hacerlo ahora algebraicamente, entonces imprimimos los resultados y tienen que dar 0 restados.

```python
from sympy import symbols, legendre, diff, simplify, Eq

# Variable simblica
x = symbols('x')
n = 3

# Definir los polinomios simblicos
Pn = legendre(n, x)
Pn1 = legendre(n + 1, x)

# Calcular la derivada de P(x)
dPn = diff(Pn, x)

# Lado izquierdo y derecho de la identidad
lhs = (1 - x**2) * dPn
rhs = (n + 1) * x * Pn - (n + 1) * Pn1

# Verificar igualdad
identity = simplify(lhs - rhs)

# Mostrar resultados
print("Lado izquierdo:")
print(lhs)
print("\nLado derecho:")
print(rhs)
print("\nSimplificacin (lhs - rhs):")
print(identity)

# Verificacin
if identity == 0:
    print("\nLa identidad se cumple algebraicamente para n = 3.")
else:
    print("\nLa identidad NO se cumple.")
```

Para el inciso e tenemos que desarrollar ambos lados de la igualdad y verificar si se cumple con el Polinomio de Legrende:

```python
from sympy import symbols, diff, factorial, legendre, simplify, expand

x = symbols('x')
#Para el polinimio de Rodrigues:
def rodrigues_formula(n):
    return (1/(2**n * factorial(n))) * diff((x**2 - 1)**n, x, n)

# Verificamos para n = 0 hasta 5:
print("Verificacin de la frmula de Rodrigues para P(x):")
for n in range(6):
    P_rodrigues = expand(rodrigues_formula(n))
    P_legendre = legendre(n, x)

    print(f"\nn = {n}:")
    print(f"Por Rodrigues: {P_rodrigues}")
    print(f"Por SymPy:     {P_legendre}")
    print(f"Coinciden?   {P_rodrigues.equals(P_legendre)}")
```
