```python
import numpy as np
import matplotlib.pyplot as plt
#Llamamos la librer√≠a para resolver los ecuaciones
x = np.linspace(0, 1, 100)

# Modos para X(0) = X(1) = 0 (senos)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
for n in range(1, 5):
    k = (n * np.pi) ** 2
    X = np.sin(n * np.pi * x)
    plt.plot(x, X, label=f"n={n}, k={k:.2f}")
plt.title("Soluciones - X(0) = X(1) = 0")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.grid(True)
plt.legend()

# Modos para X'(0) = X'(1) = 0 (cosenos)
plt.subplot(1, 2, 2)
for n in range(0, 4):  # empieza desde n=0
    k = (n * np.pi) ** 2
    X = np.cos(n * np.pi * x)
    plt.plot(x, X, label=f"n={n}, k={k:.2f}")
plt.title("Soluciones - X'(0) = X'(1) = 0")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
