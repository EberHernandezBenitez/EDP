```python
import numpy as np
import random as r
from scipy import stats
```

Usando lo último visto en clase para simular $x_1$ y $x_2$ con medias $\mu_1=1.5$ y $\mu_2=1.48$, con varianzas $\sigma_1^2=0.16$ y $\sigma_2^2=0.09$

```python
def gausiana(n , mu, sigma):
    muestra = []
    for i in range(n):
        suma = 0
        for j in range(12):
            u = r.random()
            suma += u
        x = mu + sigma*(suma-6)
        muestra.append(x)
    return muestra
```

Necesitamos ver las interferencias que se generan cuando $x_2$ supera a $x_1$

```python
def prob(n):
    numero=0
    p= 0
    x1 = gausiana(100000, 1.5,0.04)
    x2 = gausiana(100000,1.48,0.03)
    for i in range(n):
        if x1[i] < x2[i]:
            numero += 1
        p = numero/n
    return p
```

Veremos si la probabilidad se aproxima a la calculada analiticamente (34.35%)

```python
p = prob(1000)
print(p)
```

Ahora, generamos la simulacion:

```python
def sims(p, error, confianza):
    alfa = 1 - confianza
    z = stats.norm.ppf(1 - alfa / 2)
    n = (z**2 * p * (1 - p)) / (error**2)
    return float(n)
```

Usando los valores que nos da el problema:

```python
simulaciones = round(sims(p, 0.01, 0.95))
print(f"Número de simulaciones: {simulaciones}")
```

Ahora sabiendo el numero de simulaciones tenemos la mejor aproximacion:

```python
Probabilidad = prob(simulaciones)
print(f"La probabilidad de que haya interferencia es {Probabilidad*100:.2f}")
```
