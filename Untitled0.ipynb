ESTA ES UNA GRAFICA DE LA FUNCION CUADRATICA $y=x^2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data points for x and y
x = np.linspace(-10, 10, 100)  # 100 points between -10 and 10
y = x**2  # Calculate y values

# Plot the data
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafica de y= x^2")
plt.grid(False)  # Add a grid for better visualization

# Display the plot
plt.show()
```
