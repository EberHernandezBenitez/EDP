<font color="blue">ESTA ES UNA GRAFICA DE LA FUNCION CUADRATICA $y=x^2$ </font>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)  # 100 puntos entre -10 y 10
y = x**2  # funcion

# Plot the data
plt.plot(x, y,color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafica de $y= x^2$",color='cyan')
plt.grid(False)  # agregar cuadricula para una mejor visualizaci√≥n
# Display the plot
plt.show()
```
