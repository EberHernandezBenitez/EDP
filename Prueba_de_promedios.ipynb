Vamos a calcular los promedios $\bar{x}=\frac{1}{n}\sum_{i=1}^{n} x_i$\
Con $x_i$ cada elemento de los datos y $n$ el nùmero de elementos.\
Le asignamos un nivel de confianza del 95% para un buen resultado.

Calculamos la media ya que el estadistico de prueba lo requiere, asi:
$$Z_0=\frac{\bar{x}-0.5(\sqrt{n})}{\sqrt{\frac{1}{12}}}$$

```python
def prueba_promedios(xi, confianza=0.95):
    n = len(xi)
    media_x = np.mean(xi)

    z0 = (media_x - 0.5)*(np.sqrt(n)) / (np.sqrt(1/12))

    # Valor crítico Z (distribución normal estándar)
    alfa = 1 - confianza
    z_critico = stats.norm.ppf(1 - alfa/2)

    print(f"Media calculada: {media_x:.5f}")
    print(f"Estadístico Z0: {z0:.5f}")
    print(f"Valor Crítico Z_alfa/2: {z_critico:.5f}")

    if abs(z0) <= z_critico:
        print("Resultado: NO se rechaza H_0")
    else:
        print("Resultado: Se rechaza H_0")

prueba_promedios(datos)
```
